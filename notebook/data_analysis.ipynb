```python
def mysplit(s):
    head = s.rstrip('0123456789')
    tail = s[len(head):]
    return head, tail
```

```python
def mean(directory):
    lags = [256, 128, 64, 32, 16, 8, 4, 2]
    folds = ['bispectrum', 'rwb']
    recap = pd.DataFrame(index=lags, columns=folds)
    for files in os.listdir(directory):
        path = os.path.join(directory, files)
        filename, lag = mysplit(files.split('_')[1].split('.')[0])
        data = np.loadtxt(path, delimiter=",", skiprows=1, usecols=2)
        recap.loc[lag, filename] = np.mean(data)
    recap.to_csv(os.path.join(directory,'recap.csv'))
```
